```python
from pathlib import Path
import pandas as pd

# Thư mục chứa video
real_folder = Path(r"C:\Users\TanPhat\Documents\DEEPLEARNING\deepfake_detection\data\Celeb-DF-v2\Celeb-real")
fake_folder = Path(r"C:\Users\TanPhat\Documents\DEEPLEARNING\deepfake_detection\data\Celeb-DF-v2\Celeb-synthesis")

# Lấy danh sách file video với full path
real_videos = [str(real_folder / f) for f in real_folder.iterdir() if f.suffix in ['.mp4', '.avi']]
fake_videos = [str(fake_folder / f) for f in fake_folder.iterdir() if f.suffix in ['.mp4', '.avi']]

# Tạo DataFrame
df = pd.DataFrame({
    'video_path': real_videos + fake_videos,
    'label': ['Real']*len(real_videos) + ['Fake']*len(fake_videos)
})

# Lưu CSV
df.to_csv('celebrity_videos.csv', index=False)

# Hiển thị 5 dòng mẫu
print(df.sample(5))


```
