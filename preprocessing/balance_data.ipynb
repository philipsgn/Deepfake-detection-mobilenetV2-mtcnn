```python
import pandas as pd
from sklearn.utils import resample
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\TanPhat\Documents\DEEPLEARNING\deepfake_detection\preprocessing\celebrity_videos.csv')  
df.sample(5)
```

```python
df_real = df[df['label']=='Real']
df_fake = df[df['label']=='Fake']
```

```python
df_fake_resampled = resample(df_fake,
                             replace=False,
                             n_samples=1500,
                             random_state=42)
df_balanced = pd.concat([df_real, df_fake_resampled])
df_balanced.head()
```

```python
counts = df_balanced['label'].value_counts()
plt.bar(counts.index, counts.values, color=['blue','red'])
plt.title("Balanced Celeb-DF v2")
plt.show()
```

```python
df_balanced.to_csv('balanced_videos.csv', index=False)
```
